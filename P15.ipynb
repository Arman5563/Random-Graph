```python
import numpy as np
import matplotlib.pyplot as mplot
from numpy.random import default_rng
import networkx as nx

def isolation_connection(n,p,iteration):
    connected_p=0
    isolation=0
    for i in range(iteration):
        g=nx.fast_gnp_random_graph(n,p)
        nodes=g.nodes()
        p1=0
        for node in nodes:
            n1=g.neighbors(node)
            if (len(list(n1))==0):
                p1=p1+1/n
        if (p1==0):
            connected_p=connected_p+1
        isolation=isolation+p1
    return isolation/iteration,connected_p/iteration
x=np.arange(10,160,10)
connected_p1=np.array([])
isolation_p1=np.array([])
for n in x:
    a,b=isolation_connection(n,4/n,150)
    connected_p1=np.append(connected_p1,[b])
    isolation_p1=np.append(isolation_p1,[a])
fig2,ax2=mplot.subplots(figsize=(14, 4))
ax2.set_title('Results for p(n)=4/n')
mplot.plot(x,connected_p1,'c',label="connected_p")
mplot.plot(x,isolation_p1,'r',label="isolation_p")
mplot.legend()
```
