```python
import numpy as np
import matplotlib.pyplot as mplot
from numpy.random import default_rng
import networkx as nx
@np.vectorize
def mean_triangles(n,p,iteration):
    c=0
    for i in range(iteration):
        g=nx.fast_gnp_random_graph(n,p)
        nodes=g.nodes()
        for node in nodes:
            c=c+nx.triangles(g,node)
    return c/(3*iteration)
x=np.arange(10,110,10)
p=60/x**2
mean=mean_triangles(x,p,100)
fig1,ax1=mplot.subplots(figsize=(14, 4))
mplot.xticks(np.arange(0,110,10))#suitable value ticks on x axis
ax1.set_title('Average number of triangles')
mplot.xlabel('n')
mplot.plot(x,mean,label="p(n)=60/n^2")
mplot.legend()
```
