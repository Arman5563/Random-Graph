```python
import numpy as np
import matplotlib.pyplot as mplot
from numpy.random import default_rng
import networkx as nx
@np.vectorize
def mean_triangles(n,p,iteration):
    c=0
    for i in range(iteration):
        g=nx.fast_gnp_random_graph(n,p)
        nodes=g.nodes()
        for node in nodes:
            c=c+nx.triangles(g,node)
    return c/(3*iteration)
x=np.arange(50,1250,50)
p=1/x
mean=mean_triangles(x,p,100)
cumsum=0
cummean=np.zeros(x.size)
for i in range(x.size):
    cumsum=cumsum+mean[i]
    cummean[i]=cumsum/(i+1)
fig1,ax1=mplot.subplots(figsize=(14, 4))
ax1.set_title('Average number of triangles')
mplot.xlabel('n')
mplot.xticks(np.arange(0,1300,100))#suitable value ticks on x axis
mplot.plot(x,mean,'r',label="p(n)=1/n")
mplot.legend()
fig2,ax2=mplot.subplots(figsize=(14, 4))
mplot.xticks(np.arange(0,1300,100))#suitable value ticks on x axis
ax2.set_title('Cumulative mean of average number of triangles')
mplot.xlabel('n')
mplot.plot(x,cummean,'c',label="p(n)=1/n")
mplot.legend()
```
